## Exploratory Data Analysis

Exploratory data analysis (EDA) is an approach of analyzing data sets to summarize their main characteristics, using statistical graphics and other data visualization methods.

**Importance**

The main purpose of exploratory data analysis (EDA) is to help look at data before making any decision/assumptions. It can help identify obvious errors, as well as better understand patterns within the data, detect outliers or anomalous events, find interesting relations among the variables. In otherwords, it gives more insight into the data.

EDA can be conducted in the following ways:

**Univariate:**

i. Non-graphical: This is the simplest form of data analysis, where the data being analyzed consists of just one variable. Since it’s a single variable, it doesn’t deal with causes or relationships.
ii. Graphical:- Non-graphical methods don’t provide a full picture of the data. Graphical methods are therefore required. It involves visual exploratory analysis of the data.

**Multivariate:**

i. Non-graphical: Multivariate non-graphical EDA techniques generally shows the relationship between two or more variables in the data, through cross-tabulation or statistics.
ii. Graphical: Multivariate data uses graphics to display relationships between two or more sets of data. The most used graphic is a grouped bar plot or bar chart with each group representing one level of one of the variables and each bar within a group representing the levels of the other variable.

```python
# Initialize an empty string to store words from tags
comment_words = ''

# Set of stopwords to be excluded from the WordCloud
stopwords = set(STOPWORDS)

# Iterate through each tag in the 'tag' column of the 'tags' DataFrame
for val in tags['tag']:
    # Typecast each tag to a string
    val = str(val)

    # Split the tag into tokens
    tokens = val.split()

    # Convert each token to lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()

    # Join the lowercase tokens and add to the comment_words string
    comment_words += " ".join(tokens) + " "

# Generate a WordCloud using the comment_words
wordcloud = WordCloud(width=800, height=800,
                      background_color='black',
                      stopwords=stopwords,
                      min_font_size=10).generate(comment_words)

# Plot the WordCloud image
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")

# Set the title for the WordCloud plot
plt.title('Distribution of words in the tags data frame by Tags')

# Adjust layout for better spacing
plt.tight_layout(pad=0)

# Display the WordCloud plot
plt.show()
```

We are able to deduce the importance of certain words that could play a role in the functionality of our algorithem. The most used words in the tags dataset were:

* Sci-fi
* Twist Ending
* Nudity

```python
# Define a function to clean movie titles using regular expressions
# This function takes a title string as input and removes any characters that are not letters, digits, or spaces
def clean_titles(title):
    # Use the re.sub() function to substitute any character that is not a letter, digit, or space with a space
    # The regex pattern "[^a-zA-Z0-9 ]" matches any character that is not a letter (a-z, A-Z), digit (0-9), or space
    # It is replaced by a space, effectively removing special characters, punctuation, etc.
    cleaned_title = re.sub("[^a-zA-Z0-9 ]", " ", title)
    return cleaned_title

# Apply the clean_titles function to the 'title' column of the 'new_data' DataFrame
# This creates a new column 'clean_title' containing cleaned versions of the movie titles
new_data["clean_title"] = new_data["title"].apply(clean_titles)

# Display the first few rows of the 'new_data' DataFrame to show the cleaned movie titles
new_data.head()
```

```python
# This operation joins the values from multiple columns into a single column, separated by spaces
# The selected features are 'genres', 'plot_keywords', 'title_cast', and 'clean_title'
# The resulting column is named 'joined_features'
joined_features = new_data['genres'] + ' ' + new_data['plot_keywords'] + ' ' + new_data['title_cast'] + ' ' + new_data['clean_title']
```

```python
# Import the TfidfVectorizer class from the scikit-learn library
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize a TfidfVectorizer object with specified parameters
# ngram_range=(1,2) indicates that both unigrams and bigrams should be considered
# stop_words='english' specifies that common English stop words should be removed during vectorization
# This helps to filter out irrelevant words that do not contribute much to the meaning of the text
vectorizer = TfidfVectorizer(ngram_range=(1,2), stop_words='english')

# Apply the TfidfVectorizer to the 'joined_features' data
# The fit_transform() method is used to learn the vocabulary from the 'joined_features' data and transform it into a TF-IDF matrix
tfidf = vectorizer.fit_transform(joined_features)

# Display the shape of the TF-IDF matrix
# The shape attribute returns a tuple representing the dimensions of the matrix (number of rows, number of columns)
# This provides information about the size of the TF-IDF matrix, where the number of rows corresponds to the number of documents (movies) and the number of columns corresponds to the unique terms (words) in the vocabulary
tfidf.shape
```

```python
# Compute the cosine similarity matrix using linear_kernel
# The linear_kernel function computes the dot product of each pair of rows in the tfidf matrix
# This effectively calculates the cosine similarity between each pair of documents (movies)
# The resulting cosine similarity matrix represents the similarity between all pairs of movies based on their TF-IDF representations
cosine_sim = linear_kernel(tfidf, tfidf)
```

```python
# Construct a reverse map of indices and movie titles
# This creates a pandas Series where the index is the clean_title of the movies and the values are their corresponding indices
# It allows for efficient lookup of the index of a movie based on its clean_title
# The `pd.Series()` function is used to create the Series, with the movie indices as values and clean_titles as index
# The `drop_duplicates()` method ensures that each clean_title is unique in the resulting Series
indices = pd.Series(new_data.index, index=new_data['clean_title']).drop_duplicates()
```

```python
# This function takes a movie title as input and returns a list of titles of the most similar movies based on cosine similarity scores.
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the input title
    movie_index = indices[title]

    # Get the pairwise similarity scores between the input movie and all other movies in the dataset
    similarity_scores = list(enumerate(cosine_sim[movie_index]))

    # Sort the movies based on their similarity scores in descending order
    similarity_scores_sorted = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Extract the similarity scores of the 10 most similar movies (excluding the input movie itself)
    top_similar_scores = similarity_scores_sorted[1:11]

    # Extract the indices of the 10 most similar movies
    top_similar_indices = [index[0] for index in top_similar_scores]

    # Retrieve the titles of the top 10 most similar movies from the 'clean_title' column
    top_similar_movie_titles = new_data['clean_title'].iloc[top_similar_indices]

    # Return the titles of the top 10 most similar movies
    return top_similar_movie_titles
```

#### Collaborative Filtering

Collaborative filtering-based recommender systems solely rely on past interactions between users and items in order to suggest new products. The features of every individual item are not considered.

In collaborative filtering, the historical data of the user interacting with the items is recorded and stored. This is usually represented by a matrix known as user-item interaction matrix, where rows represent users and columns represent the items. Similar users are grouped and all their interactions are considered when making recommendations to the target user.

```python
# Creating a Reader object with a specified rating scale
# The rating_scale parameter defines the range of ratings in the dataset, which helps Surprise understand the rating values
# In this case, the rating scale is set from 0.5 to 5.0, indicating that ratings in the dataset range from 0.5 to 5.0
reader = Reader(rating_scale=(0.5, 5.0))
```

#### Singular Value Decomposition (SVD) Model

SVD is a method of decomposing a matrix, usually called A, into three smaller matrices: U, S, and V. U is an orthogonal matrix with unit vectors that are perpendicular to each other. It captures the direction of the data in A. S is a diagonal matrix with values on the main diagonal only.

```python
# Start measuring the time taken to train the model
modelstart = time.time()

# Create an instance of the SVD (Singular Value Decomposition) algorithm with default parameters
# SVD is a matrix factorization technique commonly used in recommender systems
# The parameters n_factors, n_epochs, lr_all, and reg_all control the model's complexity and regularization
svd = SVD(n_epochs=35,n_factors=400,init_std_dev=0.001,random_state=42,verbose=True)

# Train the SVD model on the training set
svd.fit(trainset)

# Generate predictions for the test set using the trained model
predictions = svd.test(testset)

# Calculate evaluation metrics to assess the model's performance
# Root Mean Squared Error (RMSE) measures the average magnitude of the errors between predicted and actual ratings
rmse_svd = accuracy.rmse(predictions)

# Mean Absolute Error (MAE) measures the average absolute difference between predicted and actual ratings
mae_svd = accuracy.mae(predictions)

# Mean Squared Error (MSE) measures the average of the squares of the errors between predicted and actual ratings
mse_svd = accuracy.mse(predictions)

# Print the evaluation metrics to assess the model's accuracy
print('RMSE: ', rmse_svd)
print('MAE: ', mae_svd)
print('MSE: ', mse_svd)

# Print the total runtime of the model training process
print("Model Runtime: %0.2f seconds" % ((time.time() - modelstart)))
```
